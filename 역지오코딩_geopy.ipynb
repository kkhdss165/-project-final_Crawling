## 역지오코딩 - geopy - 버스 주소

```python
from geopy.geocoders import Nominatim
import pandas as pd
import numpy as np

def geocoding_reverse(lat_lng_str):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    address = geolocoder.reverse(lat_lng_str)

    return address

data = pd.read_csv('./bus_info.csv', encoding= 'CP949', index_col= 0, header= 0, engine= 'python')

# print(data.X좌표)

xpos_s = data.X좌표
ypos_s = data.Y좌표

address_s = []
n_pos = len(xpos_s)
for n in range(0,n_pos):
    xpos = xpos_s[n]
    ypos = ypos_s[n]

    # print(xpos, ypos)
    if np.isnan(xpos) == False:
        pos = str(ypos) + ', '+str(xpos)
        # print(pos)
        address = geocoding_reverse(pos)

        # print(type(address))
        # print(dir(address))
        # print(address.address)
        print("%d/%d" % (n,n_pos) ,xpos, ypos, address)

        address_s.append(address)
    else:
        print("%d/%d" % (n, n_pos), "NONE")
        address_s.append('')

print(address_s)
```

### 주소 컬럼 추가

```python
# 예외처리후 컬럼 추가

## 예외 처리
## '•' 문자 제거
exception_texts = [u'\u2022', u'\u697c', u'\u0e19', u'\u0e49', u'\u0e33', u'\u0e21', u'\u0e31', u'\u0e2a']


address_s2 = address_s

address_col = []

for idx,address in enumerate(address_s2):
    if address != '':
        str_address = address.address
        
        for exception_text in exception_texts:
            str_address = str_address.replace(exception_text, '')
        
    else :
        str_address = ''
#     print(idx,str_address)
    address_col.append(str_address)
    
#address_col 

data = pd.read_csv('./bus_info.csv', encoding= 'CP949', index_col= 0, header= 0, engine= 'python')
data['주소'] = address_col
data.to_csv("./bus_info_address.csv", encoding="cp949", mode='w', index = True)
```
